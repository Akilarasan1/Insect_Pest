# Species from Gbif - Halyomorphoa_halys (2007-2023)

```python
# Define the "plot_species_data" function to display the selected species data within the extent of the bioclimatic background
def plot_species_data(species_data, bioclim_data):
    fig, ax = plt.subplots(figsize=(8, 6))

    # Load and plot the bioclimatic background
    with rio.open(bioclim_data, 'r') as src:
        rioplot.show(src, ax=ax, cmap="turbo")

    # Load the species data
    species_df = gpd.read_file(species_data)

    # Filter the species data to include only points within the extent of the bioclimatic background
    species_df = species_df.cx[src.bounds[0]:src.bounds[2], src.bounds[1]:src.bounds[3]]

    # Plot the filtered species data points
    species_df.plot(ax=ax, markersize=5, color='black')

    # Set the title of the plot
    ax.set_title("Annual_Mean_Temperature_WC and \nSpecies from Gbif - Halyomorphoa_halys (2007-2023)", fontsize=12)

# Define the bioclimatic data path
bioclim = '/content/drive/MyDrive/Project_2/Insect Pest/tif files/US_Clip_From_World_Clim/Bio_1_Annual_mean_Temperature_WC_1970_2000.tif'

# Define the species data file path
species_data = '/content/drive/MyDrive/Project_2/Insect Pest/Species Data/GBIF Site/BrownStinkBug(2007-2023).gpkg'

# Call the plot_species_data function to display the plot
plot_species_data(species_data, bioclim)

plt.show()
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import rasterio as rio
import rasterio.plot as rioplot

def plot_species_data(species_data1, species_data2, bioclim_data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    # Load and plot the bioclimatic background
    with rio.open(bioclim_data, 'r') as src:
        rioplot.show(src, ax=ax1, cmap="turbo")
        rioplot.show(src, ax=ax2, cmap="turbo")

    # Load and plot the species data points
    species_df1 = gpd.read_file(species_data1)
    # Filter the species data to include only points within the extent of the bioclimatic background
    species_df1 = species_df1.cx[src.bounds[0]:src.bounds[2], src.bounds[1]:src.bounds[3]]

    species_df2 = gpd.read_file(species_data2)
    species_df1.plot(ax=ax1, markersize=5, color='black')
    species_df2.plot(ax=ax2, markersize=5, color='black')

    # Set the titles of the plots
    ax1.set_title("Annual_Mean_Temperature_WC and \nSpecies from GBif - Halyomorphoa_halys (2007-2023)", fontsize=12)
    ax2.set_title("Annual_Mean_Temperature_WC and \nSpecies from EDD Maps - Halyomorpha_halys (2007-2023)", fontsize=12)

    plt.show()

# Define the bioclimatic data path
bioclim = '/content/drive/MyDrive/Project_2/Insect Pest/tif files/US_Clip_From_World_Clim/Bio_1_Annual_mean_Temperature_WC_1970_2000.tif'

# Define the species data file paths
species_data1 = '/content/drive/MyDrive/Project_2/Insect Pest/Species Data/GBIF Site/BrownStinkBug(2007-2023).gpkg'
species_data2 = '/content/drive/MyDrive/Project_2/Insect Pest//Species Data/EDD Maps site/Halyomorpha_halys(1999_2023).gpkg'

# Call the plot_species_data function to display the plots side by side
plot_species_data(species_data1, species_data2, bioclim)
```
